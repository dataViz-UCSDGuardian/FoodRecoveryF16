```python
import pandas as pd
import bokeh.charts as bk
import bokeh.plotting as plt
```

```python
df = pd.read_excel("FoodOver.xlsx", sheetname = "Sheet3")

col = [str(x) + " " + str(y) + "lbs." for (a,(x,y,w)) in df.iterrows()] 

df["Label"] = col
df
```

```python
fig = bk.Donut(df, label=['Week', 'Label'], values='lbs.',
               text_font_size='8pt', width= 900, height= 900)
plt.show(fig)
```
